Снаряд выстреливается под углом θ к горизонту с начальной скоростью v₀.

Игнорируем сопротивление воздуха.

Нужно смоделировать траекторию движения:

Формулы:

x(t)=v₀*cos(θ)*t \
y(t)=v₀*sin(θ)*t-0.5*g*t²

где g = 9.81 м/с².

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

g=9.81
N=100
data=[]

def trajectory(v0,theta):
    global g
    global N
    global data

    # рассчетные формулы
    t_end=2*v0*np.sin(theta)/g
    t=np.linspace(0,t_end,N)

    x=v0*np.cos(theta)*t
    y=v0*np.sin(theta)*t-0.5*g*t**2

    data.append({
        "Угол (°)": np.rad2deg(theta),
        "Время полета (с)": round(t_end, 2),
        "Дальность (м)": round(x[-1], 2),
        "Макс. высота (м)": round(y.max(), 2)
    })
    return x,y
    
v0=7

x_30,y_30=trajectory(v0, np.pi/6)
x_45,y_45=trajectory(v0,np.pi/4)
x_60,y_60=trajectory(v0,np.pi/3)



# построение графика
plt.figure()
plt.plot(x_30, y_30,label='Начальный угол 30')
plt.plot(x_45, y_45,label='Начальный угол 45')
plt.plot(x_60, y_60,label='Начальный угол 60')

plt.title('Траектории полета без сопротивления воздуха')
plt.xlabel('Дальность')
plt.ylabel('Высота')
plt.legend()
plt.show()

df = pd.DataFrame(data)
print(df)
```
